```python
import threading
import time
import random

# Função para Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    print(f"Bubble Sort concluído. Resultado: {arr[:10]}...")

# Função para Merge Sort
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2  # Divida o array em duas metades
        L = arr[:mid]
        R = arr[mid:]

        merge_sort(L)  # Ordene a primeira metade
        merge_sort(R)  # Ordene a segunda metade

        i = j = k = 0

        # Copie os dados temporários de volta para o array original
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        # Verifique se há algum elemento sobrando
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
    print(f"Merge Sort concluído. Resultado: {arr[:10]}...")

# Função para Selection Sort
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    print(f"Selection Sort concluído. Resultado: {arr[:10]}...")

# Função para criar uma lista aleatória
def criar_lista_aleatoria(tamanho):
    return [random.randint(0, 1000) for _ in range(tamanho)]

if __name__ == "__main__":
    threads = []
    tamanhos = [100000, 100000, 100000]  # Tamanhos das listas para cada algoritmo
    listas = [criar_lista_aleatoria(tamanho) for tamanho in tamanhos]

    # Bubble Sort
    t1 = threading.Thread(target=lambda: bubble_sort(listas[0].copy()))
    threads.append(t1)
    t1.start()

    # Merge Sort (substitui Quick Sort)
    t2 = threading.Thread(target=lambda: merge_sort(listas[1].copy()))
    threads.append(t2)
    t2.start()

    # Selection Sort (substitui Shell Sort)
    t3 = threading.Thread(target=lambda: selection_sort(listas[2].copy()))
    threads.append(t3)
    t3.start()

    for t in threads:
        t.join()

    print("Todas as threads concluídas.")
```

```python
import threading
import time
import random

# Função para Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    print(f"Bubble Sort concluído. Resultado")

# Função para Quick Sort
def quick_sort(arr):
    def partition(low, high):
        pivot = arr[high]
        i = low - 1
        for j in range(low, high):
            if arr[j] < pivot:
                i += 1
                arr[i], arr[j] = arr[j], arr[i]
        arr[i+1], arr[high] = arr[high], arr[i+1]
        return i+1

    def quick_sort_recursion(low, high):
        if low < high:
            pi = partition(low, high)
            quick_sort_recursion(low, pi-1)
            quick_sort_recursion(pi+1, high)

    quick_sort_recursion(0, len(arr) - 1)
    print(f"Quick Sort concluído. Resultado")

# Função para Shell Sort
def shell_sort(arr):
    n = len(arr)
    gap = n // 2
    while gap > 0:
        for i in range(gap, n):
            temp = arr[i]
            j = i
            while j >= gap and arr[j - gap] > temp:
                arr[j] = arr[j - gap]
                j -= gap
            arr[j] = temp
        gap //= 2
    print(f"Shell Sort concluído. Resultado")

# Função para criar uma lista aleatória
def criar_lista_aleatoria(tamanho):
    return [random.randint(0, 1000) for _ in range(tamanho)]

if __name__ == "__main__":
    threads = []
    tamanhos = [100000, 100000, 100000]  # Tamanhos das listas para cada algoritmo
    listas = [criar_lista_aleatoria(tamanho) for tamanho in tamanhos]

    # Bubble Sort
    t1 = threading.Thread(target=lambda: bubble_sort(listas[0].copy()))
    threads.append(t1)
    t1.start()

    # Quick Sort
    t2 = threading.Thread(target=lambda: quick_sort(listas[1].copy()))
    threads.append(t2)
    t2.start()

    # Shell Sort
    t3 = threading.Thread(target=lambda: shell_sort(listas[2].copy()))
    threads.append(t3)
    t3.start()

    for t in threads:
        t.join()

    print("Todas as threads concluídas.")
```
